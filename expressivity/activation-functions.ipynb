# Imports

```python
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
%matplotlib inline
```

# Helper functions

```python
def plot_module(module, xmin=-3, xmax=3, h=.1):
    x = torch.arange(xmin, xmax+h, h, requires_grad=True)
    y = module.forward(x)
    y.backward(torch.ones_like(x))

    fig, ax = plt.subplots(figsize=(6, 4))

    ax.plot(x.detach().numpy(), y.detach().numpy(), label=f"y = f(x)")
    ax.plot(x.detach().numpy(), x.grad.detach().numpy(), label="y = f'(x)")
    ax.set_xlim(xmin, xmax)
    ax.grid();
    ax.legend();
    ax.set_title(f"{module._get_name()}", fontsize=22)
    return fig, ax

# Test
plot_module(nn.ReLU());
```

# `nn.ReLU`

```python
module = nn.ReLU()
fig, ax = plot_module(module)
ax.set_title("ReLU", fontsize=22, y=1.07);
```

# `nn.LeakyReLU`

```python
module = nn.LeakyReLU(negative_slope=.2)
fig, ax = plot_module(module, xmin=-5, xmax=5)
ax.set_title("LeakyReLU", fontsize=22, y=1.07);
```

# `nn.ELU`

```python
module = nn.ELU()
fig, ax = plot_module(module, xmin=-5, xmax=5)
ax.set_title("ELU", fontsize=22, y=1.07);
```

# `nn.Tanh`

```python
module = nn.Tanh()
fig, ax = plot_module(module)
ax.set_title("Tanh", fontsize=22, y=1.07);
```

# `nn.Sigmoid`

```python
module = nn.Sigmoid()
fig, ax = plot_module(module, xmin=-5, xmax=5)
ax.set_title("Sigmoid", fontsize=22, y=1.07);
```
